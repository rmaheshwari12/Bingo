```python
import pandas as pd
import numpy as np
import random
import string
import time
from IPython.display import display, HTML
```

```python
def bingo_sheet(numrange,sheets):
    '''
    Get the required bingo sheets for your group !
    '''
    i = 0
    while i <= sheets:
        bingo = pd.DataFrame(np.random.choice(numrange,size=(5, 5),replace =False), columns=list('BINGO'));
        bingo = bingo.style.set_properties(**{
        'background-color': 'lightblue', #you can change color of the background here
        'color' : 'black', #and font color here
        'font-size': '25pt',
        'title' : '25pt'})
        i += 1
        print('\n')
        display('Sheet no. {}'.format(i),bingo)
        print('\n')
    return 0


def bingo_pick(interval, numrange):
    while True:
        print(random.choice('BINGO'),np.random.randint(0,numrange))
        time.sleep(interval)
```

```python
bingo_sheet(numrange=50, sheets=10)
```

# Ready, Set, BINGO ! 

```python
bingo_pick(interval=10, numrange=50)
```
